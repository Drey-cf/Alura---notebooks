**As operações matemáticas em arrays Numpy são muito mais rápidas do que em listas regulares, pois a NumPy foi otimizada para trabalhar com conjuntos de dados homogêneos e libera memória do computador rapidamente.**

```python
import numpy as np
import time

# cria uma lista com 1000000 elementos
lista = list(range(1000000))

# transforma a lista em um array Numpy
array = np.array(lista)

# começa a cronometrar o tempo para a operação com a lista
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento da lista
lista_quadrado = [i**2 for i in lista]

# para o cronômetro
list_time = time.time() - start_time

# começa a cronometrar o tempo para a operação com o array
start_time = time.time()

# realiza a operação de elevar ao quadrado cada elemento do array
array_quadrado = array**2

# para o cronômetro
array_time = time.time() - start_time

print("Tempo da operação com a lista: ", list_time)
print("Tempo da operação com o array: ", array_time)
```

# <p style = 'text-align : center; color : red'> Orange vs Grapefruit

```python
df = np.loadtxt('citrus.csv', delimiter = ',', skiprows = 1, usecols =np.arange(1, 6, 1) )
df
```

```python
df.shape
```

```python
dl = df[:5000, 0]
pl = df[:5000, 1]
dt = df[5000:, 0]
pt = df[5000:, 1]
```

```python
plt.plot(dl, pl)
plt.plot(dt, pt)

plt.show()
```

```python
x = dl             #para laranjas
y = pl             #definindo as váriaveis pro calculo de al e bl
n = np.size(dl)

al = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - (np.sum(x))**2) #calculando, por minimos quadrados al e
bl = np.mean(y) - al*np.mean(x)              #bl

plt.plot(dl, pl)
plt.plot(x, al*x + bl )
plt.show()
```

```python
x = dt             #para toranjas
y = pt             #definindo as váriaveis pro calculo de at e bt
n = np.size(dt)

at = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - (np.sum(x))**2) #calculando, por minimos quadrados al e
bt = np.mean(y) - at*np.mean(x)                                              #bl

plt.plot(dt, pt)
plt.plot(x, at*x + bt )
plt.show()
```
